## Analyzing Adults with advanced education 
 

##### First task is to analyze the data. For adults with advanced education, earning over 50K

```python
advanced_ed_mask = (adult_df.education == 'Bachelors') | (adult_df.education == 'Masters') | (adult_df.education=='Doctorate')
```

```python
adult_df[advanced_ed_mask] # adults with advanced education
```

```python
adult_df[~(advanced_ed_mask)] # adults without advanced education
```

```python
# percentage of adults with higher education, earning over 50K 
adult_df[advanced_ed_mask].salary.value_counts(normalize=True) * 100
```

#### `46.5%` of adults with advanced education, earn 50K+ monthly.

### Adults without advanced education.

```python
# ~(advanced_ed_mask) stands for all adults without Bachelors, Masters or a Doctorate.
adult_df[~(advanced_ed_mask)].salary.value_counts(normalize=True) * 100
```

#### `17.4%` of adults without advanced education earn over 50K monthly.

### Salary Analysis

```python
# We will be analyzing based on salary.
# Adults earning above 50K, they are considered rich. 
# Salaries below or equal to 50K are not in the rich-category
```

```python
salary_grouped = adult_df.groupby('salary')
rich_adults = salary_grouped.get_group('>50K')
```

### Adults earning 50K+ salary

```python
rich_adults
```

```python
rich_adults['education'].value_counts(normalize=True) * 100
```

```python
(rich_adults['education'].value_counts(normalize=True) * 100)[['Bachelors', 'Masters', 'Doctorate']].sum()
```

#### `44.5%` of adults earning 50K+ in salary, have `advanced education`.

```python
100 - (rich_adults['education'].value_counts(normalize=True) * 100)[['Bachelors', 'Masters', 'Doctorate']].sum()
```

#### `55.5%` of adults earning 50K+ in salary, do not have `advanced education`.
